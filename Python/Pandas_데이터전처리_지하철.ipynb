## 01. 1월 - 5월 전처리

```python
# 같은 컬럼을 가진 [0:5]를 묶어주기
first_df = pd.concat(total[0:5])
```

```python
first_df
```

### 01-1) 알고 싶은 정보들만 모아서 전처리

```python
# masking 기법 활용
first_df[(first_df.날짜 == '2016-01-01') & (first_df.역명 == '서울역(150)') & (first_df['구 분'] == '승차')].sum()
```

```python
# query method활용
first_df.query(""" 날짜 == '2016-01-01' and 역명 == '서울역(150)' """)
```

```python
# 데이터를 하나로 합치는 과정
first_2 = first_df.groupby(['날짜', '호선', '역명', '구 분'], as_index=False).sum()
```

### 01-2) 역명 수정

```python
# '('를 기반으로 역명 이름과 뒤 숫자를 나눠주고, apply로 역명 이름만 채용
first_2['역명'] = first_2.역명.apply(lambda x : x.split('(')[0])
```

```python
first_3 = first_2.groupby(['날짜', '역명', '구 분']).sum()
```

```python
first_3
```

```python
first_3.reset_index(inplace=True)
```

```python
first_3
```

```python
first_3.columns = [x.replace(" ", "") for x in first_3.columns]
```

```python
first_3.drop(['02~03', '03~04'], axis=1, inplace=True) # axis=1 : column제거
```

```python
first_3['역명'].value_counts() # 역마다 304개의 데이터가 존재하는 것을 확인. 
```

## 02. 5월 ~ 11월

```python
second_df = pd.concat(total[5:-1])
```

```python
import re
```

```python
p = re.compile("[가-힣]+")
```

```python
second_df['역명'] = second_df['역명'].apply(lambda x : p.findall(x)[0])
```

```python
second_2 = second_df.groupby(['역명', '날짜', '구분'], as_index = False).sum()
```

```python
second_2.columns
```

```python
# 앞서 전처리한 first_3와 비교
# 합집함 - 여집합 => 겹치지 않는 부분 확인
(set(second_2.columns) | set(first_3.columns)) - (set(second_2.columns) & set(first_3.columns))
```

```python
subway_1 = pd.concat([second_2, first_3]).drop(['04~05'], axis=1) # axis=1 : 축 기준 drop
```

## 03. 12월 

```python
# 12월에 대해서 시간에 대한 '시' 없애주기
total[-1].columns = [x.replace("시", "") for x in total[-1].columns]
```

```python
# 역명 바꿔주기
total[-1]['역명'] = total[-1]['역명'].apply(lambda x : p.findall(x)[0])
```

```python
total[-1] = total[-1].groupby(['날짜', '역명', '구분'], as_index=False).sum()
```
