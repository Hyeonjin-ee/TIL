# matplotlib API 개요

- 정보 시각화는 데이터 분석에서 무척 중요
- 시각화는 특잇값을 찾아내거나, 데이터 변형이 필요한지 알아보거나, 모델에 대한 아이디어를 찾기 위한 과정의 일부
- 파이썬 시각화 도구로 matplotlib 활용

- 공식 사이트
https://matplotlib.org/

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
plt.plot([1, 2, 3], [2, 4, 6]) # 각 각의 리스트가 x축, y축으로 전달되어 그려진 선 그래프
plt.title("Hello matplotlib")
plt.show() # 주피터에서는 plt.show()를 안해줘도 그림이 표시되지만, .py 같은 스크립트 작성 시에는 필요
````

### Figure와 Axes 

**Figure**
- 그림을 그리기 위한 도화지의 역할. 그림판의 크기 등을 조절함

**Axes**
- 실제 그림을 그리는 메소드들과, x축, y축, title 등의 속성을 가짐

```
plt.plot([1, 2, 3], [2, 4, 6])  # 기본 설정값의 Figure가 만들어진 후 내부적으로 Axes.plot() 호출
plt.title("Hello plot")         # Axes.set_title() 호출
plt.show()                      # Figure.show() 호출
```

````python
plt.figure(figsize=(10,5)) # 안써주면 default 크기로 생성
plt.plot([1, 2, 3], [2, 4, 6]) 
plt.title("Hello matplotlib")
plt.show() 
````

````python
# Figure의 색깔을 조정
plt.figure(facecolor='red') # 생성된 Figure 객체에 색깔을 입힘
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Hello plot")
plt.show()
````

````python
# Figure를 가져옴
plt.figure()
````

````python
# Axes를 가져옴
plt.axes()
````

````python
# 여러 개의 plot를 가지는 figure 설정

fig, ax = plt.subplots()
print(type(fig), type(ax))
````

### plot 함수 안에 올 수 있는 데이터

### 색상, 마커, 라인스타일 등 변경

````python
# 색상 변경
plt.plot([1, 2, 3], [1, 2, 3], color="green")
````

### x축, y축에 축명을 텍스트로 할당. xlabel, ylabel 적용

````python
plt.plot([1, 2, 3], [1, 2, 3], color="green", marker='o', linestyle='dashed', linewidth=3, markersize=10 )
plt.xlabel("x axis")
plt.ylabel("y axis")
````

### X축, Y축 틱값 표기

````python
plt.plot([1, 2, 3], [1, 2, 3], color="green", marker='o', linestyle='dashed', linewidth=3, markersize=10 )
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.xticks([1, 2, 3]) # x축 눈금 지정
plt.yticks([1, 2, 3]) # y축 눈금 지정
plt.show()
````

### x축, y축 틱값을 표현을 회전해서 보여줌
- x축값이 문자열이고 많은 tick값이 있을 때 적용 

````python
plt.plot(["one","two","three"],[1, 2, 3], [1, 2, 3], color="skyblue", marker='o', linestyle='dashed', linewidth=2, markersize=10 )
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.xticks(rotation=30) # x축 눈금 지정
plt.yticks() # y축 눈금 지정
plt.show()
````

### xlim()은 x축값을 제한하고, ylim()은 y축값을 제한

````python
x = np.arange(100)
y = x*2

plt.plot(x,y, color="skyblue")
plt.xticks(np.arange(0, 100, 10))
plt.show()
````

### 범례 설정하기

````python
x = np.arange(100)
y = x*2

plt.plot(x, y, label="test", color="skyblue")              
plt.xticks(np.arange(0, 100, 5))

# x축과 y축에 표시될 구간 설정
plt.xlim(0, 50)
plt.ylim(0, 100)

plt.legend()
plt.show()
````

### 여러개의 plot을 하나의 Axes내에서 그리기

````python
x1 = np.arange(100)
y1 = x*2

x2 = np.arange(100)
y2 = x*3


plt.plot(x1, y1, color = "green", label="y=2x")         
plt.plot(x2, y2, color = "red", label="y=3x")

plt.xticks(np.arange(0, 100, 5))

# x축과 y축에 표시될 구간 설정
plt.xlim(0, 50)
plt.ylim(0, 100)

plt.legend()
plt.show()
````

````python
x1 = np.arange(100)
y1 = x*2

plt.plot(x1, y1, color="red", label="line")  # 선그래프   
plt.bar(x1, y1, color="skyblue", label="bar")  # 막대그래프

plt.legend()
plt.show()
````

````python
x1 = np.arange(10)
y1 = x1*2

plt.plot(x1, y1, color="red", marker='o', label="line")  # 선그래프   
plt.bar(x1, y1, color="skyblue", label="bar")  # 막대그래프

plt.xticks(ticks=x1)

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.legend()
plt.show()
````

### Axes 객체에서 직접 작업 하기

````python
figure = plt.figure()
ax = plt.axes()

ax.plot(x1, y1, color="red", marker='o', label="line")
ax.bar(x1, y1, color="skyblue", label="bar")

ax.set_xticks(ticks=x1)

ax.set_xlabel("x axis")
ax.set_ylabel("y axis")

ax.legend()
plt.show()
````

### 여러개의 subplots을 가지는 Figure를 생성하고 여기에 개별 그래프를 시각화

````python
x1 = np.arange(1,10)
y1 = 2 * x1

x2 = np.arange(1,20)
y2 = 2 * x2

fig, ax = plt.subplots(2,2, figsize=(12,6))

ax[0][0].plot(x1, y1, color='red', marker='o', linestyle="dashed", linewidth=2, markersize=6 , label="red line")
ax[0][1].bar(x2, y2, color='skyblue', label='blue bar')
ax[1][0].plot(x1, y1, color='skyblue', marker='o', linestyle="dashed", linewidth=2, markersize=6 , label="blue line")
ax[1][1].bar(x2, y2, color='red', label='red bar')

ax[0][0].set_xlabel("axes[0][0] x axis")
ax[0][1].set_xlabel("axes[0][1] x axis")
ax[1][0].set_xlabel("axes[1][0] x axis")
ax[1][1].set_xlabel("axes[1][1] x axis")

ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
````
